# 20.4 - Read a video stream from webcam (frame by frame)

```python
# video is series of frames and each frame is a image
```

```python
cap = cv2.VideoCapture(0)  # cv2.VideoCapture(id):- captures frames, and
# here 'id' means id of the device, id 0 means default webcam i.e default
# webcam of your device

while True:
    ret, frame = cap.read() # cap.read():- return two values, 1st is
    # boolean and 2nd is frame that are captured
    # if boolean value is false, it means frame is not captured properly,
    # and it can be due webcam is not availabe, webcam is not yet started
    # etc.
    
    if ret == False:
        continue
        
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # cv2.cvtColor(image_objcet, cv2.COLOR_BGR2GRAY):- to switch b/w two
    # color spaces
    # cv2.COLOR_BGR2GRAY:- to switch from BGR to GRAY
        
    cv2.imshow("Video Frame", frame)
    cv2.imshow("Gray Frame", gray_frame)
    
    # wait for user input 'q', then loop will stop
    key_pressed = cv2.waitKey(1) & 0xFF # here 1, means program will wait
    # for 1 ms before next iteration comes up
    # cv2.waitKey() returns 32 bit integer and 0xFF is a 8 bit number of
    # 1s i.e. (1111 1111), so when and (&) operation is performed b/w them
    # 8 bit integer is generated
    
    if key_pressed == ord('q'):  # ord():- return ASCII value of character
    # add is of 8 bit (0 - 255)
        break

cap.release()  # release():- release the device after the capture is over
cv2.destroyAllWindows()
```

# 20.5

```python
# Haarcascade clasifier
# classifier that is already trained on lots of facial data
```

```python
import cv2

cap = cv2.VideoCapture(0)

face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
# cv2.CascadeClassifier("haarcascade_frontalface_alt.xml"):- creates
# classifier object which work on facial data

while True:
    ret, frame = cap.read()
    
    if ret == False:
        continue
        
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray_frame, 1.3, 5)
    # detectMultiScale(image_object, scaleFactor, minNeighbors):- returns
    # coordinates (starting point (x, y), width (w), height (h))
    # [(x, y, w, h)] consisting of list of tuples of faces (x, y, w, h)
    
    # scaleFactor :- parameter specifies how much the image size is reduced
    # at each image scale.
    
    # minNeighbors :- parameter specifies how many neighbors each candidate
    # rectange sould have to retain it. This paramter will affect the
    # quality of the detected faces, Higher value results in less
    # detections but with the higher quality, 3 to 6 is a good value.
    
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

    cv2.imshow("Video Frame", frame)
    cv2.imshow("Gray Frame", gray_frame)
    
    key_pressed = cv2.waitKey(1) & 0xFF
    
    if key_pressed == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# 20.6

```python
# Write a pyhton script that captures images from your webcam video stream
# Extract all faces from the image frame (using haarcascades)
# Stores the face information into numpy arrays
```

```python
# 1. Read and show video stream, capture images
# 2. Detect faces and show bounding box (using haarcascades)
# 3. Flatten (converting 2D array to 1D array) the largest face image
   # (gray scale) and save in a numpy array
# 4. Repeat the above for mutiple people to generate training data
```

```python
import cv2
import numpy as np

# initialize camera
cap = cv2.VideoCapture(0)

# face detection
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")

skip = 0
face_data = []
dataset_path = './data/'
face_section = 0

file_name = input("Enter the name of the person: ")

while True:
    ret, frame = cap.read()
    
    if ret == False:
        continue
        
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
    faces = face_cascade.detectMultiScale(frame, 1.3, 5)
    faces = sorted(faces, key = lambda f : f[2] * f[3])  # sorting faces
    # according to face area
    
    # pick last face, as it is the largest face acc. to area (f[2]*f[3])
    for face in faces[-1:]:
        x, y, w, h = face
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # extract (crop out the required face) : Region of Interest
        offset = 20  # to provide padding of 10px around boundary box
        face_section = frame[y - offset : y + h + offset,
                              x - offset : x + w + offset]
        face_section = cv2.resize(face_section, (100, 100))  
        # cv2.resize(image_object, (resolution)):- resizing face section
        # to 100x100

        skip += 1
        if skip % 7 == 0:
            face_data.append(face_section)
            print(len(face_data))
    
    cv2.imshow("Frame", frame)
    cv2.imshow("Face Section", face_section)
    
    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
        
# converting our face list array into a numpy array
face_data = np.asarray(face_data)  # np.asarray():- convert list to
# numpy array 
face_data = face_data.reshape((face_data.shape[0], -1))
print(face_data.shape)

# save this data into file system
np.save(dataset_path + file_name + '.npy', face_data)
# np.save(path +qq file_name + 'file_extension', data) :- save file of
# given extension with specified name at specified path
print("Data Successfully save at " + dataset_path + file_name + '.npy')
```

# 20.7

```python
# recognise faces using some classification algorithm like logistic, KNN
# SVM etc.

# 1. load the training data (numpy arrays of all the persons)
    # x - values are stored in the numpy arrays
    # y - values we need to assign for each person
# 2. read a video stream using opencv
# 3. extract faces out of it (testing)
# 4. use knn to find the prediction of face (int)
# 5. map the predicted id to name of the user
# 6. display the predictions on the screen - bounding box and name
```

```python
import cv2
import numpy as np
import os

# ========== KNN code ==========

def distance(v1, v2):
    # Eucledian distance
    return np.sqrt(((v1-v2)**2).sum())

def knn(train, test, k=5):
    dist = []

    for i in range(train.shape[0]):
        # Get the vector and label
        ix = train[i, :-1]
        iy = train[i, -1]
        
        # Compute the distance from test point
        d = distance(test, ix)
        dist.append([d, iy])
        
    # Sort based on distance and get top k
    dk = sorted(dist, key = lambda x: x[0])[:k]
    
    # Retrieve only the labels
    labels = np.array(dk)[:, -1]
    
    # Get frequencies of each label
    output = np.unique(labels, return_counts=True)
    
    # Find max frequency and corresponding label
    index = np.argmax(output[1])
    
    return output[0][index]

# ========== Data Preparation ==========

dataset_path = './data/'

face_data = []
labels = []

class_id = 0  # lables for the given file
names = {}  # dictionary for mappig b/w id and name

for fx in os.listdir(dataset_path):  # os.listdir(path):- list all the
# files present is that specified path
    if fx.endswith('.npy'):
        names[class_id] = fx[: -4]  # mapping b/w class_id and name
        print("loaded " + fx)
        data_item = np.load(dataset_path + fx)
        face_data.append(data_item)
        
    # creating labels for the class
    target = class_id*np.ones((data_item.shape[0], ))
    class_id += 1
    labels.append(target)
         
face_dataset = np.concatenate(face_data, axis = 0)
face_labels = np.concatenate(labels, axis = 0)
face_labels = face_labels.reshape(-1, 1)

print(face_dataset.shape)
print(face_labels.shape)

trainset = np.concatenate((face_dataset, face_labels), axis = 1)
print(trainset.shape)

# ========== testing ==========

# initialize camera
cap = cv2.VideoCapture(0)

# face detection
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")

while True:
    ret, frame = cap.read()
    
    if ret == False:
        continue
    
    faces = face_cascade.detectMultiScale(frame, 1.3, 5)
    
    if (len(faces) == 0):
        continue
    
    for face in faces:
        x , y, w, h = face
        
        # getting the ROI (Region of Interest)
        offset = 10
        face_section = frame[y - offset : y + h + offset,
                              x - offset : x + w + offset]
        face_section = cv2.resize(face_section, (100, 100))
        
        # predicting label
        out = knn(trainset, face_section.flatten())  # flatten():- convert
        # frame matrix as one array
        
        # displaying name and boundary box around the face on screen
        pred_name = names[int(out)]
        cv2.putText(frame, pred_name, (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2,cv2.LINE_AA)
        # cv2.putText(image_object):-
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 255), 2)
        
    cv2.imshow("Faces", frame)

    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```
